```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_csv ('melb_data.csv')
df.head()
```

**NOMOR 1**

```python
plt.hist(df['Price'])
plt.title('Distribusi Harga Rumah', size=15)
```

**NOMOR 2**

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='Distance', y='Regionname' )
plt.title('Distribusi Jarak ke CBD berdasarkan Nama Region', size=15)
plt.xlabel('Distance')
plt.ylabel('Regionname')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**NOMOR 3**

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Type', y='Price', estimator=np.median)
plt.title('Perbandingan Nilai Median Harga Rumah', size=15)
plt.xlabel('Type')
plt.ylabel('price')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**NOMOR 4**

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Price', y='Distance', hue='Type', alpha=1)
plt.title('Korelasi Harga Rumah dan Jarak Rumah ke CBD', fontsize=15)
plt.xlabel('Harga Rumah (Price)')
plt.ylabel('Jarak ke CBD (Distance)')
plt.legend(title='Tipe Rumah')
plt.tight_layout()
plt.show()
```

**NOMOR 5**

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x='Month', y='Price', hue='Type', palette='Set1')
plt.title('Tren Harga Rumah per Bulan berdasarkan Tipe Rumah', size=15)
plt.xlabel('bulan(Month)')
plt.ylabel('Harga Rumah (Price)')
plt.legend(title='Tipe Rumah')
plt.tight_layout()
plt.show()
```
